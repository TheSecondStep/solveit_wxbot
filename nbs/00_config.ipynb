# Config

> Environment variables and constants for solveit_wxbot.

```
#| hide
from nbdev.showdoc import *
```

所有凭据均从环境变量读取，避免硬编码敏感信息。启动前需在容器或 `.env` 中设置以下变量：

| 变量名 | 说明 |
|---|---|
| `WECOM_CORP_ID` | 企业 ID |
| `WECOM_SECRET` | 应用 Secret |
| `WECOM_AGENT_ID` | 应用 Agent ID |
| `WECOM_TOKEN` | 回调 Token |
| `WECOM_AES_KEY` | 消息加解密密钥（43位，无需末尾 `=`）|

```
#| export
import base64, os

# WeCom credentials
CORP_ID  = os.environ.get('WECOM_CORP_ID', '')    # Enterprise ID
SECRET   = os.environ.get('WECOM_SECRET', '')     # App secret
AGENT_ID = os.environ.get('WECOM_AGENT_ID', '')  # App agent ID
TOKEN    = os.environ.get('WECOM_TOKEN', '')      # Webhook token

# AES key requires padding to valid base64 length before decoding
AES_KEY = base64.b64decode(os.environ.get('WECOM_AES_KEY', '') + '=')

# Tuneable constants
CHUNK_SIZE   = 680  # Max chars per WeCom message segment
DEDUP_WINDOW = 60   # Seconds to remember seen MsgIds
TOKEN_BUFFER = 300  # Seconds before expiry to refresh access token
```
